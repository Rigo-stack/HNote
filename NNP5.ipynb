```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import tensorflow as tf
import os
import cv2
```

```python
mnist = tf.keras.datasets.mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data() # load data into training and testing
```

```python
x_train = tf.keras.utils.normalize(x_train, axis=1) # normalize the data
x_test = tf.keras.utils.normalize(x_test, axis=1)
```

```python
# model = tf.keras.sequential([
#    tf.keras.layers.Flatten(input_shape=(28,28))
#    tf.keras.layers.Dense(128, activation='relu')
#    tf.keras.layers.Dense(10)
# ])
# model.compile(
#     optimizer='adam',
#     loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
#     metrics=['accuracy']
# )
```

```python
model = tf.keras.models.Sequential() # change this model to the SCRATCH  model
model.add(tf.keras.layers.Flatten(input_shape = (28,28))) # flatten model
model.add(tf.keras.layers.Dense(128, activation='relu')) # activation function
model.add(tf.keras.layers.Dense(128, activation='relu'))
model.add(tf.keras.layers.Dense(10,activation = 'softmax'))
model.compile(optimizer = 'adam',loss = 'sparse_categorical_crossentropy', metrics =['accuracy'])

model.fit(x_train, y_train, epochs = 3)

model.save('handwritten.model.keras')
```

```python
loss, accuracy = model.evaluate(x_test, y_test)
```

```python
print(loss, accuracy)
```

```python
metrics = ['Loss', 'Accuracy']
results = [loss, accuracy]

plt.figure(figsize=(8, 4))
plt.barh(metrics, results, color=['skyblue', 'lightgreen'])
plt.xlabel('Score')
plt.title('Model Evaluation Metrics')
plt.xlim(0, max(results) * 1.1)
plt.show()
```
